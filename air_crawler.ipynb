```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
url="http://www.weather.go.kr/weather/forecast/mid-term_01.jsp"
def weather_crawler(url):
    html = urlopen(url)
    bs = BeautifulSoup(html.read(), "html.parser")
    day = bs.find_all("th", class_="top_line")
    title = []
    for i in day:
        title.append(i.text)
    title1 = title[:9]
    title2 = title[10:]
    task = bs.select('table > tbody > tr > td > ul > li')
    data = []
    t = 0
    for i in task:
        if t%2 == 0:
            if i.get('title') != None:
                data.append(i.get('title'))
        t+=1
    local = bs.select('table > tbody > tr > th')
    data2 = bs.select('table > tbody > tr > th')
    second = []
    for i in data2:
        second.append(i.text)

    second = second[11:]
    second.remove("강원도영서")
    second.remove("강원도영동")
    second.remove("대전ㆍ세종ㆍ충청남도")
    second.remove("충청북도")
    second.remove("광주ㆍ전라남도")
    second.remove("전라북도")
    second.remove("부산ㆍ울산ㆍ경상남도")
    second.remove("대구ㆍ경상북도")
    second.remove("제주도")

    ddd = bs.select('td > ul > li')
    last = []
    for t in ddd:
        last.append(t.text)
    last = last[260:]
    f = open('air.csv', 'w', newline="")
    writer = csv.writer(f)
    writer.writerow(title1)
    for i in range(10):
        date = data[i*13:(i+1)*13]
        date.remove(date[1])
        date.remove(date[2])
        date.remove(date[3])
        date.remove(date[4])
        date.remove(date[5])
        writer.writerow([local[i].text]+date)
    writer.writerow(title2)
    for i in range(24):
        writer.writerow([second[i]]+last[i*8:(i+1)*8])
    f.close()
weather_crawler(url)
```
